## 1. Create an interactive bar chart showing total quantity and revenue by country (excluding United Kingdom) for the month of April 2011.

```python
import plotly.express as px


filtered_data = df[(df['InvoiceDate'].dt.month == 4) & (df['InvoiceDate'].dt.year == 2011) & (df['Country'] != 'United Kingdom')]

grouped_data = filtered_data.groupby('Country').agg({'Quantity': 'sum', 'Revenue': 'sum'}).reset_index()

fig = px.bar(grouped_data, x='Country', y=['Quantity', 'Revenue'], title='Quantity and Revenue by Country (April 2011)',
             labels={'Quantity': 'Total Quantity', 'Revenue': 'Total Revenue'}, barmode='group')

fig.show()
```

## 2. Create an interactive line chart showing quantity and revenue sold to France between January 1st and May 31st 2011.

```python
filtered_data2 = df[((df['InvoiceDate'].dt.month == 1) & (df['InvoiceDate'].dt.year == 2011)) | ((df['InvoiceDate'].dt.month == 5) & (df['InvoiceDate'].dt.year == 2011)) & (df['Country'] == 'France')]

grouped_data2 = filtered_data2[filtered_data2['Country'] == 'France'].groupby('Country').agg({'Quantity': 'sum', 'Revenue': 'sum'}).reset_index()

fig2 = px.bar(grouped_data2, x='Country', y=['Quantity', 'Revenue'], title='Quantity and Revenue by Country (Jan and May 2011)',
              labels={'Quantity': 'Total Quantity', 'Revenue': 'Total Revenue'}, barmode='group')

# Show the interactive chart
fig2.show()
```

## 3. Create an interactive scatter plot showing the relationship between average quantity (x-axis) and average unit price (y-axis) for the product PARTY BUNTING with the plot points color-coded by country (categories).

```python
#Todos los items con Party Bunting
search_term = 'PARTY BUNTING'

df_partyB= df[df['Description'].str.contains(search_term, case=False)]

df_partyB
```

```python
df_partyB

averages_by_country = df_partyB.groupby('Country').agg({'Quantity': 'mean', 'UnitPrice': 'mean'}).reset_index()

# Create an interactive scatter plot using Plotly Express
fig = px.scatter(averages_by_country, x='Quantity', y='UnitPrice', text='Country',
                 title='Average Quantity vs. Average Unit Price for PARTY BUNTING by Country',
                 labels={'Quantity': 'Average Quantity', 'UnitPrice': 'Average Unit Price'})

# Show the interactive scatter plot
fig.update_traces(textposition='top center')
fig.show()
```
